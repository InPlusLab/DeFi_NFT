## 柱状图：x轴协议，y轴交易手续费总和，柱子按4类分4个颜色

```python
tempMarketMap={}
with open("/mnt/sde1/peilin_defi/data/ethereumProfit_v1/dict/defiNftFee.map", "rb") as tf:
    tempMarketMap=pickle.load(tf)

marketMap={}
for key,value in tempMarketMap.items():
    marketMap[key]=value["fee"]/pow(10,18)
```

```python
exchange_defi_list=["UniswapV1","UniswapV2","UniswapV3", "BalancerV1","BalancerV2", "Curve", "SushiSwap", "ShibaSwap"]
value0=[marketMap[marketType] for marketType in exchange_defi_list]
reverseMap={}
for marketType in exchange_defi_list:
    reverseMap[marketMap[marketType]]=marketType
value0.sort(reverse=False)
exchange_defi_list=[reverseMap[tempValue] for tempValue in value0]

exchange_nft_list=["OpenSea", "LooksRare", "X2Y2", "Blur"]
value1=[marketMap[marketType] for marketType in exchange_nft_list]
reverseMap={}
for marketType in exchange_nft_list:
    reverseMap[marketMap[marketType]]=marketType
value1.sort(reverse=False)
exchange_nft_list=[reverseMap[tempValue] for tempValue in value1]

lending_defi_list=["AaveV1","AaveV2", "MakerDAO", "Compound", "Cream"]
value2=[marketMap[marketType] for marketType in lending_defi_list]
reverseMap={}
for marketType in lending_defi_list:
    reverseMap[marketMap[marketType]]=marketType
value2.sort(reverse=False)
lending_defi_list=[reverseMap[tempValue] for tempValue in value2]


lending_nft_list=["NFTfiV1","NFTfiV2", "Arcade", "BendDAO"]
value3=[marketMap[marketType] for marketType in lending_nft_list]
reverseMap={}
for marketType in lending_nft_list:
    reverseMap[marketMap[marketType]]=marketType
value3.sort(reverse=False)
lending_nft_list=[reverseMap[tempValue] for tempValue in value3]
```

```python
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        width = rect.get_width()
        plt.annotate('{:.0f}'.format(width),
                    xy=(width, rect.get_y() + rect.get_height() / 2),
                    xytext=(3, 0),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='left', va='center', fontsize=10)

rects3=plt.barh(lending_nft_list,value3,label="NFT Lending")
rects2=plt.barh(lending_defi_list,value2,label="Defi Lending")
rects1=plt.barh(exchange_nft_list,value1,label="NFT Exchange")
rects0=plt.barh(exchange_defi_list,value0,label="Defi Exchange")

autolabel(rects0)
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

handles, labels = plt.gca().get_legend_handles_labels()
handles = [handles[3],handles[2],handles[1],handles[0]]
labels = [labels[3],labels[2],labels[1],labels[0]]

plt.legend(handles,labels,loc=4, fontsize=16)

plt.grid(True, color='#666666', linestyle = ":", linewidth = "1",axis='x')
plt.xlabel('Total Transaction Fees (Ether)', fontsize=20)
plt.ylabel('Decentralized Markets', fontsize=20)
plt.xticks(fontsize=20)

plt.xlim(1,10000000000)
plt.xscale('log')
plt.yticks(fontsize=10)

plt.savefig("/mnt/sde1/peilin_defi/data/ethereumProfit_v1/photo/totalFee_bar.pdf",bbox_inches='tight')
```

## 饼状图：以太坊总交易手续费中，20个协议的比例（一笔交易同时涉及N个协议就除以N）（和上面这个到时估计2选1）

```python
import numpy as np
import csv    #加载csv包便于读取csv文件
import zipfile
import os
import json
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
tempMarketMap={}
with open("/mnt/sde1/peilin_defi/data/ethereumProfit_v1/dict/defiNftFee.map", "rb") as tf:
    tempMarketMap=pickle.load(tf)

marketMap={}
for key,value in tempMarketMap.items():
    marketMap[key]=value["averageFee"]/pow(10,18)
```

```python
# 排序
reverseMap={}
for key,value in marketMap.items():
    reverseMap[value]=key
    
valueList=list(marketMap.values())
valueList.sort(reverse=False)
print(valueList)

keyList=[]
for item in valueList:
    keyList.append(reverseMap[item])

print(keyList)

print(marketMap)

exchange_defi_list=["UniswapV1","UniswapV2","UniswapV3", "BalancerV1","BalancerV2", "Curve", "SushiSwap", "ShibaSwap"]
exchange_nft_list=["OpenSea", "LooksRare", "X2Y2", "Blur"]
lending_defi_list=["AaveV1","AaveV2", "MakerDAO", "Compound", "Cream"]
lending_nft_list=["NFTfiV1","NFTfiV2", "Arcade", "BendDAO"]
rgb_list=[]
for name in keyList:
    if name in exchange_defi_list:
        rgb_list.append('#1F77B4')
        continue
    
    if name in exchange_nft_list:
        rgb_list.append('#FF7F0e')
        continue
    
    if name in lending_defi_list:
        rgb_list.append('#2CA02C')
        continue
    
    if name in lending_nft_list:
        rgb_list.append('#D62728')
        continue
    
print(rgb_list)
print(len(rgb_list)==len(keyList))
```

```python
num1=1
num2=1
num3=2
num4=0

keys = list(keyList)
values = np.array(valueList)
# patches, texts = plt.pie(values,colors=rgb_list)

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        # 同时显示数值和占比的饼图
        # return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
        if pct<8.97083505988121:
            return ''
        else:
            return '{p:.2f}%'.format(p=pct)
    
    
    return my_autopct
patches, texts,_ = plt.pie(values,autopct=make_autopct(values))


# labels = ['{} {:1.2f}% ({})'.format(i, j, v) for i, j, v in
#             zip(keys, 100. * values / values.sum(), values)]  # 构造图例数据
labels = ['{} ({:1.2f}%) '.format(i, j) for i, j in
            zip(keys, 100. * values / values.sum())]  # 构造图例数据

# labels=keys

patches, labels, dummy = zip(*sorted(
    zip(patches, labels, values), key=lambda x: x[2], reverse=True))

# plt.legend(patches[:10], labels[:10], fontsize=10,bbox_to_anchor=(num1, num2), loc=num3, borderaxespad=num4)
plt.legend(patches, labels, fontsize=10,bbox_to_anchor=(num1, num2), loc=num3, borderaxespad=num4,ncol=2)
plt.tight_layout()
plt.savefig("/mnt/sde1/peilin_defi/data/ethereumProfit_v1/photo/feeTotal.pdf",bbox_inches='tight')
```

## 柱状图：x轴协议，y轴平均单笔交易手续费，柱子按4类分4个颜色

```python
tempMarketMap={}
with open("/mnt/sde1/peilin_defi/data/ethereumProfit_v1/dict/defiNftFee.map", "rb") as tf:
    tempMarketMap=pickle.load(tf)

marketMap={}
for key,value in tempMarketMap.items():
    marketMap[key]=float(value["fee"]/pow(10,18)/value["count"])
```

```python
exchange_defi_list=["UniswapV1","UniswapV2","UniswapV3", "BalancerV1","BalancerV2", "Curve", "SushiSwap", "ShibaSwap"]
value0=[marketMap[marketType] for marketType in exchange_defi_list]
reverseMap={}
for marketType in exchange_defi_list:
    reverseMap[marketMap[marketType]]=marketType
value0.sort(reverse=False)
exchange_defi_list=[reverseMap[tempValue] for tempValue in value0]

exchange_nft_list=["OpenSea", "LooksRare", "X2Y2", "Blur"]
value1=[marketMap[marketType] for marketType in exchange_nft_list]
reverseMap={}
for marketType in exchange_nft_list:
    reverseMap[marketMap[marketType]]=marketType
value1.sort(reverse=False)
exchange_nft_list=[reverseMap[tempValue] for tempValue in value1]

lending_defi_list=["AaveV1","AaveV2", "MakerDAO", "Compound", "Cream"]
value2=[marketMap[marketType] for marketType in lending_defi_list]
reverseMap={}
for marketType in lending_defi_list:
    reverseMap[marketMap[marketType]]=marketType
value2.sort(reverse=False)
lending_defi_list=[reverseMap[tempValue] for tempValue in value2]


lending_nft_list=["NFTfiV1","NFTfiV2", "Arcade", "BendDAO"]
value3=[marketMap[marketType] for marketType in lending_nft_list]
reverseMap={}
for marketType in lending_nft_list:
    reverseMap[marketMap[marketType]]=marketType
value3.sort(reverse=False)
lending_nft_list=[reverseMap[tempValue] for tempValue in value3]
```

```python
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        width = rect.get_width()
        plt.annotate('{:.04f}'.format(width),
                    xy=(width, rect.get_y() + rect.get_height() / 2),
                    xytext=(3, 0),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='left', va='center', fontsize=10)

rects3=plt.barh(lending_nft_list,value3,label="NFT Lending")
rects2=plt.barh(lending_defi_list,value2,label="Defi Lending")
rects1=plt.barh(exchange_nft_list,value1,label="NFT Exchange")
rects0=plt.barh(exchange_defi_list,value0,label="Defi Exchange")

autolabel(rects0)
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

handles, labels = plt.gca().get_legend_handles_labels()
handles = [handles[3],handles[2],handles[1],handles[0]]
labels = [labels[3],labels[2],labels[1],labels[0]]

plt.legend(handles,labels,loc=1, fontsize=16)

plt.grid(True, color='#666666', linestyle = ":", linewidth = "1",axis='x')
plt.xlabel('Average Transaction Fees (Ether)', fontsize=20)
plt.ylabel('Decentralized Markets', fontsize=20)
plt.xticks(fontsize=15)
plt.xlim(0,0.08)
plt.yticks(fontsize=10)

plt.savefig("/mnt/sde1/peilin_defi/data/ethereumProfit_v1/photo/averageFee_bar.pdf",bbox_inches='tight')
```
